```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
```

```python
image = Image.open('../data_original/HAM10000_images_part_1/ISIC_0024306.jpg')
```

```python
_ = plt.imshow(image)
```

```python
desired_size = max(image.size)
desired_size
```

```python
old_size = image.size
```

```python
delta_w = desired_size - old_size[0]
```

```python
delta_h = desired_size - old_size[1]
```

```python
padding = (delta_w // 2, delta_h // 2, delta_w - (delta_w // 2), delta_h - (delta_h // 2))
```

```python
resized_image = ImageOps.expand(image, padding)
```

```python
_ = plt.imshow(resized_image)
```

```python
resized_image = resized_image.resize((224, 224))
# resized_image.resize((224, 224))
```

```python
_ = plt.imshow(resized_image)
```
